```python
import numpy as np
import time
import matplotlib.pyplot as plot
from collections import deque


# INSERTION SORT

def insertion_Sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j] :
            arr[j + 1] = arr[j]
            j =j-1
        arr[j + 1] = key
    return arr




# MERGE SORT

def merge(arr, l, m, r):
    arr1size = m - l + 1
    arr2size = r - m

    LArray = [0] * (arr1size)
    RArray = [0] * (arr2size)
    for i in range(0, arr1size):
        LArray[i] = arr[l + i]
    for j in range(0, arr2size):
        RArray[j] = arr[m + 1 + j]
    i = 0
    j = 0
    k = l
    while i < arr1size and j < arr2size:
        if LArray[i] <= RArray[j]:
            arr[k] = LArray[i]
            i += 1
        else:
            arr[k] = RArray[j]
            j += 1
        k += 1
    while i < arr1size:
        arr[k] = LArray[i]
        k += 1
        i += 1
    while j < arr2size:
        arr[k] = RArray[j]
        k += 1
        j += 1



def merge_Sort(arr, l, r):
    if l < r:
        m = l+(r-l)//2
        merge_Sort(arr, l, m)
        merge_Sort(arr, m+1, r)
        merge(arr, l, m, r)



# QUICK SORT

def partition(arr, l, r):
    i = ( l - 1 )
    x = arr[r]
    for j in range(l, r):
        if   arr[j] <= x:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[r] = arr[r], arr[i + 1]
    return (i + 1)

def Quicksort_Iterative(arr, l, r):
    size = r - l + 1
    stack = [0] * (size)
    top = -1
    top = top + 1
    stack[top] = l
    top = top + 1
    stack[top] = r
    while top >= 0:
        r = stack[top]
        top = top - 1
        l = stack[top]
        top = top - 1
        v = partition( arr, l, r )
        if v-1 > l:
            top = top + 1
            stack[top] = l
            top = top + 1
            stack[top] = v - 1
        if v + 1 < r:
            top = top + 1
            stack[top] = v + 1
            top = top + 1
            stack[top] = r


# HEAP SORT

def beginHeapSort(arr):
    n = len(arr)
    for index in range(n//2 - 1, -1, -1):
        heapSortUtil(arr, index, n)
    for index in range(n-1, 0, -1):
        arr[index], arr[0] = arr[0], arr[index]
        heapSortUtil(arr, 0, index)

def heapSortUtil(arr, index, n):
    vr = index
    r = 2 * index + 2
    l = 2 * index + 1
    if r < n and arr[vr] < arr[r]:
        vr = r
    if l < n and arr[vr] < arr[l]:
        lar = l
    if vr != index:
        arr[index], arr[vr] = arr[vr], arr[index]
        heapSortUtil(arr, vr, n)


# MODIFIED QUICK SORT

def swap (Arr, i, j):
    temp = Arr[i]
    Arr[i] = Arr[j]
    Arr[j] = temp

def modifiedPartition(l, start, end):
    PivotIndex = end
    if ((end - start) > 2):
        PivotIndex = (start + end) // 2
        tempList = [l[start],l[PivotIndex],l[end]]
        maxNum = max(tempList)
        l[end] = maxNum
        tempList.remove(maxNum)
        maxNum = max(tempList)
        l[PivotIndex] = maxNum
        tempList.remove(maxNum)
        l[start] = tempList[0]
    pivot = l[PivotIndex]
    l.remove(pivot)
    PivotInd = start
    for i in range(start, end):
        if l[i] <= pivot:
            swap(l, i, PivotInd)
            PivotInd = PivotInd + 1
    l.insert(PivotInd, pivot)
    return PivotInd

def iterativeQuicksort(l):
    if (len(l)> 10):
        stack = deque()
        start = 0
        end = len(l) - 1
        stack.append((start, end))
        while stack:
            start, end = stack.pop()
            pivot = modifiedPartition(l, start, end)
            if pivot - 1 > start:
                stack.append((start, pivot - 1))
            if pivot + 1 < end:
                stack.append((pivot + 1, end))
    else:
        insertion_Sort(l)
def plotRuntime(n,insertion_Output,merge_Output,quicksort_Output,heapSort_Output,modified_QSOutput):
    plot.plot(n,insertion_Output,label="Insertion Sort" )
    plot.plot(n, merge_Output,label="Merge Sort" )
    plot.plot(n, quicksort_Output,label="Quick Sort" )
    plot.plot(n, heapSort_Output,label="Heap Sort" )
    plot.plot(n, modified_QSOutput,label="Modified Quick Sort" )
    plot.ylabel('Runtime in seconds')
    plot.legend()
    plot.show()




n = [1000, 2000, 3000, 5000, 10000, 20000, 30000, 40000, 50000, 60000]

insertion_Output=[]
merge_Output=[]
quicksort_Output=[]
heapSort_Output=[]
modified_QSOutput=[]



print("Random Value Case: ")
for val in n:
    print("\n")
    print("Output for sorting a random array of length ", val)
    np.random.seed(2021)
    array = np.random.randint(val, size=(val))


    print("\n")
    print("INSERSTION SORT")
    Start_Time = time.time()
    insertion_Sort(array)
    insertion_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    np.random.seed(2021)
    array = np.random.randint(val, size=(val))
    print("\n")
    print("MERGE SORT")
    Start_Time = time.time()
    merge_Sort(array, 0, val-1)
    merge_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    np.random.seed(2021)
    array = np.random.randint(val, size=(val))

    print("\n")
    print("QUICK SORT")
    Start_Time = time.time()
    Quicksort_Iterative(array, 0, val-1)
    quicksort_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    np.random.seed(2021)
    array = np.random.randint(val, size=(val))
    print("\n")
    print("HEAP SORT")
    Start_Time = time.time()
    beginHeapSort(array)
    heapSort_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    np.random.seed(2021)
    array = np.random.randint(val, size=(val)).tolist()
    print("\n")

    print("MODIFIED QUICK SORT")
    Start_Time = time.time()
    iterativeQuicksort(array)
    modified_QSOutput.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))
    print("\n")



print(insertion_Output)
print(merge_Output)
print(quicksort_Output)
print(heapSort_Output)
print(modified_QSOutput)

plotRuntime(n,insertion_Output,merge_Output,quicksort_Output,heapSort_Output,modified_QSOutput)


insertion_Output=[]
merge_Output=[]
quicksort_Output=[]
heapSort_Output=[]
modified_QSOutput=[]

print("\n")
print("Best Case: ")
for val in n:
    print("\n")
    print("Output for sorting a sorted array (ascending) of length ", val)
    array = [i for i in range(1,val+1)]

    print("\n")
    print("INSERSION SORT")
    Start_Time = time.time()
    insertion_Sort(array)
    insertion_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    array = [i for i in range(1,val+1)]
    print("\n")
    print("MERGE SORT")
    Start_Time = time.time()
    merge_Sort(array, 0, val-1)
    merge_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))

    array = [i for i in range(1,val+1)]
    print("\n")
    print("QUICK SORT")
    Start_Time = time.time()
    Quicksort_Iterative(array, 0, val-1)
    quicksort_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    array = [i for i in range(1,val+1)]
    print("\n")
    print("HEAP SORT")
    Start_Time = time.time()
    beginHeapSort(array)
    heapSort_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    array = [i for i in range(1,val+1)]
    print("\n")
    print("MODIFIED QUICK SORT")
    Start_Time = time.time()
    iterativeQuicksort(array)
    modified_QSOutput.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))
    print("\n")



print(insertion_Output)
print(merge_Output)
print(quicksort_Output)
print(heapSort_Output)
print(modified_QSOutput)

plotRuntime(n,insertion_Output,merge_Output,quicksort_Output,heapSort_Output,modified_QSOutput)


insertion_Output=[]
merge_Output=[]
quicksort_Output=[]
heapSort_Output=[]
modified_QSOutput=[]

print("\n")
print("Worst Case: ")
for val in n:
    print("\n")
    print("Output for sorting a reverse-sorted array of length ", val)
    array = [i for i in range(val,0,-1)]

    print("\n")
    print("INSERSTION SORT")
    Start_Time = time.time()
    insertion_Sort(array)
    insertion_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))

    array = [i for i in range(val,0,-1)]
    print("\n")
    print("MERGE SORT")
    Start_Time = time.time()
    merge_Sort(array, 0, val-1)
    merge_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    array = [i for i in range(val,0,-1)]
    print("\n")
    print("QUICK SORT")
    Start_Time = time.time()
    Quicksort_Iterative(array, 0, val-1)
    quicksort_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    array = [i for i in range(val,0,-1)]
    print("\n")
    print("HEAP SORT")
    Start_Time = time.time()
    beginHeapSort(array)
    heapSort_Output.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))


    array = [i for i in range(val,0,-1)]
    print("\n")
    print("MODIFIED QUICK SORT")
    Start_Time = time.time()
    iterativeQuicksort(array)
    modified_QSOutput.append((time.time() - Start_Time))
    print("--- %s seconds ---" % (time.time() - Start_Time))
    print("\n")



print(insertion_Output)
print(merge_Output)
print(quicksort_Output)
print(heapSort_Output)
print(modified_QSOutput)

plotRuntime(n,insertion_Output,merge_Output,quicksort_Output,heapSort_Output,modified_QSOutput)





```
